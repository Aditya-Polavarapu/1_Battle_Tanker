```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig
from peft import PeftModel, PeftConfig
import torch
import gc

def cleanup():
    if 'model' in globals(): del globals()['model']
    if 'tokenizer' in globals(): del globals()['tokenizer']
    if 'finance_agent' in globals(): del globals()['finance_agent']
    gc.collect()
    torch.cuda.empty_cache()

cleanup()

model_id = "Adit251/DeepSeek-R1-Distill-Llama-Finance-8B"

# 1. Load the configuration to find the base model
config = PeftConfig.from_pretrained(model_id)
base_model_name = config.base_model_name_or_path

# 2. Setup Quantization
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True
)

print(f"Loading base model: {base_model_name}")
tokenizer = AutoTokenizer.from_pretrained(model_id)

# 3. Load Base Model
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_name,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True
)

# 4. Load the Adapter (DeepSeek-R1-Distill-Llama-Finance-8B)
print(f"Loading adapter: {model_id}")
model = PeftModel.from_pretrained(base_model, model_id)

# 5. Create Pipeline
globals()['finance_agent'] = pipeline("text-generation", model=model, tokenizer=tokenizer)
print("Finance Agent Pipeline successfully initialized.")
```

```python
def get_ai_signal(row):
    prompt = f"""Analyze the following stock data:
    Stock: {row['ShareName']}
    Price: {row['Close']}
    RSI: {row['Current RSI']}
    Trend: {row['Trend']}

    Based on these technical indicators, provide a trading signal (BUY, SELL, or HOLD) and a confidence probability (0-100).
    Format: Signal: [SIGNAL], Probability: [PROBABILITY]
    """
    # Short generation to extract specific signal and probability
    output = finance_agent(prompt, max_new_tokens=50, do_sample=False)
    return output[0]['generated_text']

# Example usage on the combined dataframe
# combined_df['AI_Analysis'] = combined_df.apply(get_ai_signal, axis=1)
```

```python
import sys
import os

# Deep clean of pandas from sys.modules
for x in list(sys.modules.keys()):
    if x.startswith('pandas'):
        del sys.modules[x]

import pandas as pd
import numpy as np
from concurrent.futures import ThreadPoolExecutor

# 1. Update paths for Colab environment
input_dir = '/content/15_Min'
output_dir = '/content/Indicators'
os.makedirs(output_dir, exist_ok=True)
output_excel_file = "/content/Analysis_of_Stocks.xlsx"
final_output_file = "/content/Final_Result_AI.xlsx"

# --- Core Logic from Analyser(1) ---
def smoothrng(df, period, multiplier):
    df['EMA'] = df['Close'].ewm(span=period, adjust=False).mean()
    df['AbsDiff'] = df['Close'].diff().abs()
    df['AvgRange'] = df['AbsDiff'].ewm(span=period, adjust=False).mean()
    df['SmoothRange'] = df['AvgRange'].ewm(span=(period * 2 - 1), adjust=False).mean() * multiplier
    return df

def run_ai_analysis(df):
    results = []
    print(f"Processing {len(df.groupby('ShareName'))} stocks with AI...")
    for name, group in df.groupby('ShareName'):
        last_row = group.iloc[-1]
        try:
            ai_response = get_ai_signal(last_row)
            results.append({
                'ShareName': name,
                'DateTime': last_row['DateTime'],
                'Close': last_row['Close'],
                'RSI': last_row['Current RSI'],
                'Trend': last_row['Trend'],
                'AI_Analysis': ai_response
            })
        except Exception as e:
            print(f"Error processing {name}: {e}")

    ai_df = pd.DataFrame(results)
    ai_df.to_excel(final_output_file, index=False)
    print(f"AI Analysis complete. Saved to {final_output_file}")
    return ai_df
```

```python
import glob

def process_files_and_run_ai():
    all_files = glob.glob(os.path.join(input_dir, "*.xlsx"))
    print(f"Found {len(all_files)} files.")

    all_data = []
    for f in all_files[:5]: # Processing first 5 as a test to manage time/resources
        try:
            df = pd.read_excel(f)
            df['ShareName'] = os.path.basename(f).replace('_15min.xlsx', '')
            # Apply indicators (simplification of original logic)
            df = smoothrng(df, 50, 3.0)
            df['Current RSI'] = 50 # Placeholder if RSI logic isn't fully copied
            df['Trend'] = 'Neutral'
            all_data.append(df)
        except Exception as e:
            print(f"Error reading {f}: {e}")

    if all_data:
        master_df = pd.concat(all_data)
        ai_results = run_ai_analysis(master_df)
        return ai_results
    return None

# Run the analysis
final_results = process_files_and_run_ai()
```

```python
import glob
import pandas as pd
import numpy as np
import os
from IPython.display import display

def final_production_analysis(limit=99):
    all_files = glob.glob(os.path.join(input_dir, "*.xlsx"))[:limit]
    print(f"Running Final AI Alignment for {len(all_files)} stocks...")

    all_results = []

    for f in all_files:
        try:
            name = os.path.basename(f).replace('_15min.xlsx', '')
            df = pd.read_excel(f)
            df.columns = [str(col).strip() for col in df.columns]

            close_col = next((c for c in df.columns if 'close' in c.lower()), 'Close')
            date_col = next((c for c in df.columns if 'date' in c.lower()), 'DateTime')

            df = df.rename(columns={close_col: 'Close', date_col: 'DateTime'})
            df['DateTime'] = pd.to_datetime(df['DateTime'])
            df = smoothrng(df, 50, 3.0)

            # RSI Calculation matching rule-based logic
            delta = df['Close'].diff()
            gain = (delta.where(delta > 0, 0)).rolling(14).mean()
            loss = (-delta.where(delta < 0, 0)).rolling(14).mean()
            df['RSI'] = (100 - (100 / (1 + (gain / loss)))).fillna(50)
            df['Trend'] = np.where(df['Close'] > df['EMA'], 'Up', 'Down')

            last_rows = df.tail(5)
            context = last_rows[['DateTime', 'Close', 'RSI', 'Trend']].to_string(index=False)
            last_row = last_rows.iloc[-1]

            # Rigid prompt template to stop hallucinations and garbage tokens
            prompt = f"Instruction: You are a financial expert. Analyze the 15-minute candle sequence provided below and determine the intraday signal for {name}.\n\nData Sequence:\n{context}\n\nRequired Output Format:\nReasoning: [1 sentence logic]\nSignal: [BUY/SELL/HOLD]\nProbability: [0-100]%"

            # Use explicit decoding parameters to clean output
            output = finance_agent(prompt, max_new_tokens=150, do_sample=False, truncation=True)
            gen_text = output[0]['generated_text'].split('Required Output Format:')[-1].strip()

            all_results.append({
                'ShareName': name,
                'DateTime': last_row['DateTime'].strftime('%Y-%m-%d %H:%M'),
                'Close': last_row['Close'],
                'RSI': round(last_row['RSI'], 2),
                'Trend': last_row['Trend'],
                'AI_Analysis': gen_text
            })
        except Exception as e:
            print(f"Error processing {name}: {e}")

    final_df = pd.DataFrame(all_results)
    final_df.to_excel(final_output_file, index=False)
    print("Analysis completed with refined AI logic.")
    display(final_df.head(20))
    return final_df

final_df = final_production_analysis(99)
```

```python
def debug_ai_reasoning(stock_name='RELIANCE'):
    file_path = f'/content/15_Min/{stock_name}_15min.xlsx'
    if not os.path.exists(file_path):
        print(f'File {file_path} not found.')
        return

    df = pd.read_excel(file_path)
    df.columns = [str(col).strip() for col in df.columns]
    close_col = next((c for c in df.columns if 'close' in c.lower()), 'Close')
    df = df.rename(columns={close_col: 'Close'})
    df = smoothrng(df, 50, 3.0)

    # Recalculate indicators for the test
    delta = df['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(14).mean()
    df['RSI'] = (100 - (100 / (1 + (gain / loss)))).fillna(50)
    df['Trend'] = np.where(df['Close'] > df['EMA'], 'Up', 'Down')

    recent = df.tail(10) # Give more context
    context = recent[['Close', 'RSI', 'Trend']].to_string(index=False)

    # Explicit Reasoning Prompt
    prompt = f"""Detailed Technical Analysis Task:
Stock: {stock_name}
Data Context (Last 10 intervals):
{context}

Instructions:
1. Observe the trend of RSI and Price.
2. Determine if there is a Buy/Sell/Hold opportunity based on intraday momentum.
3. Output your reasoning first, then the final Signal and Probability.

Format:
Reasoning: [Your logic]
Signal: [BUY/SELL/HOLD]
Probability: [0-100]"""

    print(f"--- Diagnostic Run for {stock_name} ---")
    output = finance_agent(prompt, max_new_tokens=300, do_sample=True, temperature=0.7)
    print(output[0]['generated_text'])

debug_ai_reasoning()
```
